#changes required
1) contest 4 to 10 have float values, need to be converted to int
2) missing values from contest 4 to 10 need to be handled

```python
backup=ratings.copy()
```

```python
def fill_with_row_mean(row, columns):
    row_mean = row[columns].mean()
    if pd.notna(row_mean):
        if row_mean.is_integer():
            row_mean = int(row_mean)
        else:
            return row.astype(float).fillna(row_mean)
        return row.fillna(row_mean)
    else:
        return row

# Specify the columns to consider for the mean calculation
columns_to_consider = ['contest1', 'contest2', 'contest3', 'contest4', 'contest5', 'contest6', 'contest7', 'contest8', 'contest9', 'contest10']

# Apply the function to each row
ratings = ratings.apply(lambda row: fill_with_row_mean(row, columns_to_consider), axis=1)
```

```python
ratings.info()
```

```python
ratings.sample(10)
```

```python
for i in range(1, 11):
    feature = 'contest' + str(i)
    ratings[feature] = ratings[feature].apply(lambda x: int(x) if not pd.isna(x) else x).astype('Int64')
ratings['rating'] = ratings['rating'].apply(lambda x: int(x) if not pd.isna(x) else x).astype('Int64')
```

```python
for i in range(1,11):
    feature = 'contest' + str(i)
    ratings=ratings[ratings[feature]>1000].reset_index(drop=True)
ratings.info()
```
