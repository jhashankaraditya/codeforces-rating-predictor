```python
import pandas as pd
import numpy as np
```

```python
ratings=pd.read_csv('cleaned_regression_df.csv')
```

```python
X=ratings.drop(columns='rating')
y=ratings['rating']
```

```python
X
```

```python
y
```

```python
X.drop(columns=['Unnamed: 0'],inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr=lr.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, X, y, cv=5, scoring='r2')
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", np.mean(scores))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
rf_gs = RandomForestRegressor(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rf_gs, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
print("Best Parameters:", best_params)
print("Best R^2 Score:", best_score)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param_dist = {
    'n_estimators': randint(100, 500),
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': randint(2, 11),
    'min_samples_leaf': randint(1, 5),
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
rf_rs = RandomForestRegressor(random_state=42)
```

```python
random_search = RandomizedSearchCV(estimator=rf_rs, param_distributions=param_dist, n_iter=100, cv=5, scoring='r2', n_jobs=-1, random_state=42)
```

```python
random_search.fit(X_train, y_train)
```

```python
best_params = random_search.best_params_
best_score = random_search.best_score_
```

```python
print("Best Parameters:", best_params)
print("Best R^2 Score:", best_score)
```

```python
from sklearn.svm import SVR
svm_regressor1 = SVR(kernel='linear')
svm_regressor2 = SVR(kernel='rbf')
```

```python
svm_regressor1.fit(X_train, y_train)
svm_regressor2.fit(X_train, y_train)
```

```python
y_pred1 = svm_regressor1.predict(X_test)
y_pred2 = svm_regressor2.predict(X_test)
```

```python
linear_r2 = r2_score(y_test, y_pred1)
rbf_r2 = r2_score(y_test, y_pred2)
print("R^2 Score Linear:", linear_r2)
print("R^2 Score RBF:", rbf_r2)
```
